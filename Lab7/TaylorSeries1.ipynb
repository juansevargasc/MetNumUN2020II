**Function**
* e ^ 2

```python
def func_e_to_2(n):
    x = 2
    e_to_2 = 0
    
    for i in range(n):
        e_to_2 += x**i/math.factorial(i)
    
    return e_to_2
```

```python
out = func_e_to_2(10)
print("e ^ 2: ", out)
```

**Function - More general**
* e ^ x

```python
def func_e(x, n):
    e_approx = 0
    for i in range(n):
        e_approx += x**i/math.factorial(i)
    
    return e_approx
```

```python
out = func_e(2,10)
print(out)
```

**Comparison**
* e ^ 5

```python
out = func_e(5,10)
print('e ^ 5 - Taylor: ', out)

out = math.exp(5)
print('e ^ 5 - Library:', out)
```

**Error calculations**

```python
x = 5
for i in range(1,11):
    e_approx = func_e(x,i)
    e_exp = math.exp(x)
    e_error = abs(e_approx - e_exp)
    print(f'{i} terms: Taylor Series approx= {e_approx}, exp calc= {e_exp}, error = {e_error}')
```

**Exit a loop in a determined error range**

```python
x = 5
for i in range(1,20):
    e_approx = func_e(x,i)
    e_exp = math.exp(x)
    e_error = abs(e_approx - e_exp)
    if e_error < 1:
        break
        
print(f'{i} terms: Taylor Series approx= {e_approx}, exp calc= {e_exp}, error = {e_error}')
```

**Cosine aproximation**

```python
def func_cos(x, n):
    cos_approx = 0
    for i in range(n):
        coef = (-1)**i
        num = x**(2*i)
        denom = math.factorial(2*i)
        cos_approx += ( coef ) * ( (num)/(denom) )
    
    return cos_approx
```

```python
angle_rad = (math.radians(45))
out = func_cos(angle_rad,5)
print(out)
```

```python
out = math.cos(angle_rad)
print(out)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# if using a Jupyter notebook, include:
%matplotlib inline

angles = np.arange(-2*np.pi,2*np.pi,0.1)
p_cos = np.cos(angles)
t_cos = [func_cos(angle,3) for angle in angles]

fig, ax = plt.subplots()
ax.plot(angles,p_cos)
ax.plot(angles,t_cos)
ax.set_ylim([-5,5])
ax.legend(['cos() function','Taylor Series - 3 terms'])

plt.show()
```

```python
angles = np.arange(-2*np.pi,2*np.pi,0.1)
p_cos = np.cos(angles)

fig, ax = plt.subplots()
ax.plot(angles,p_cos)

# add lines for between 1 and 6 terms in the Taylor Series
for i in range(1,6):
    t_cos = [func_cos(angle,i) for angle in angles]
    ax.plot(angles,t_cos)

ax.set_ylim([-7,4])

# set up legend
legend_lst = ['cos() function']
for i in range(1,6):
    legend_lst.append(f'Taylor Series - {i} terms')
ax.legend(legend_lst, loc=3)

plt.show()
```

* **sin(x)**

```python
def func_sin(x, n):
    approx = 0
    for i in range(n):
        coef = (-1)**i
        num = x ** ( 2*i + 1 )
        denom = math.factorial( 2*i + 1 )
        approx += ( coef ) * ( (num)/(denom) )
        
    return approx
```

```python
x = (math.radians(45)); n = 20
print('x = ', x /(math.pi), 'pi ' , ' n = ', n)

out = func_sin(x , n)
print('sin x - Taylor: ', out)

out = math.sin(x)
print('sin x - Library:', out)
```

```python
%matplotlib inline

angles = np.arange(-2*np.pi,2*np.pi,0.1)
taylor = np.sin(angles)
function = [func_sin(angle, 5) for angle in angles]

fig, ax = plt.subplots()
ax.plot(angles,taylor)
ax.plot(angles,function)
ax.set_ylim([-5,5])
ax.legend(['sin() function','Taylor Aprox.'])

plt.show()
```

* **1 / (1 - x)**

```python
def func_1_x(x, n):
    approx = 0
    for i in range(n):
        
        approx += ( (x ** i) )
        
    return approx
```

```python
x = math.pi / 4; n = 20
print('x = ', x, ' n = ', n)

out = func_1_x(x , n)
print('1 / 1- x - Taylor: ', out)

out = 1 / (1 - x)
print('1 / 1 - x - Python:', out)
```

```python
%matplotlib inline

x_values = np.arange(-1, 1, 0.05)
lib_ex = [ ( 1 / (1-x) ) for x in x_values]
ex = [func_1_x(x, 10) for x in x_values]

fig, ax = plt.subplots()
ax.plot(x_values, lib_ex)
ax.plot(x_values, ex)
ax.set_ylim([-1,2])
ax.legend(['1/(1-x) function','Taylor Aprox.'])

plt.show()
```

* **ln (1 + x)**

```python
def func_ln_x1(x, n):
    approx = 0
    for i in range(n):
        coef = (-1) ** (( i + 1) - 1)
        num = x ** ( i + 1 )
        denom = i + 1
        approx += ( coef ) * ( (num)/(denom) )
        
    return approx
```

```python
x = math.pi / 5; n = 20
print('x = ', x, ' n = ', n)

out = func_ln_x1(x, n)
print('ln(x+1) - Taylor: ', out)

out = math.log(1 + x)
print('ln(x+1) - Library:', out)
```

```python
x = 0.1
for i in range(1,11):
    e_approx = func_ln_x1(x,i)
    e_exp = math.log(x+1)
    e_error = abs(e_approx - e_exp)
    print(f'{i} terms: Taylor Series approx= {e_approx}, exp calc= {e_exp}, error = {e_error}')
```

```python
%matplotlib inline

x_values = np.arange(-1, 1, 0.1)
lib_ex = np.log(x_values + 1)
ex = [func_ln_x1(x, 5) for x in x_values]

fig, ax = plt.subplots()
ax.plot(x_values, lib_ex)
ax.plot(x_values, ex)
ax.set_ylim([-1, 1])
ax.legend(['ln( x+1 ) function','Taylor Aprox.'])

plt.show()
```

* **ln (x)**

```python
def func_ln_x(x, n):
    approx = 0
    for i in range(n):
        coef = (-1) ** (( i + 1) - 1)
        num = (x - 1) ** ( i + 1 )
        denom = i + 1
        approx += ( coef ) * ( (num)/(denom) )
        
    return approx
```

```python
x = math.pi / 6; n = 20
print('x = ', x, ' n = ', n)

out = func_ln_x(x, n)
print('1 / 1 - x - Taylor: ', out)

out = math.log(x)
print('1 / 1 - x - Library:', out)
```

```python
%matplotlib inline

x_values = np.arange(-1, 1, 0.1)
lib_ex = np.log(x_values)
ex = [func_ln_x(x, 3) for x in x_values]

fig, ax = plt.subplots()
ax.plot(x_values, lib_ex)
ax.plot(x_values, ex)
ax.set_ylim([-1, 1])
ax.legend(['ln(x) function','Taylor Aprox.'])

plt.show()
```
